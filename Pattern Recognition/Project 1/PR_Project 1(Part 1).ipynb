https://stackoverflow.com/questions/12851791/removing-numbers-from-string

```python
import re

def cleanhtml(file):
  cleanr = re.compile("(<.*?>)|(\.)|(\;)|(\:)|(\!)|(\')|(\?)|(\,)|(\")|(\()|(\))|(\[)|(\])")
  cleantext = re.sub(cleanr, '', file)
  cleantext =''.join([i for i in cleantext if not i.isdigit()])
  return cleantext
```
